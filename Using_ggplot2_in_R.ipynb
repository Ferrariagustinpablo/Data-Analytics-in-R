## geom_point + geom_segment + geom_text

```
head(gapminder)
```

```
# Filter
gm2007 <- gapminder %>%
  filter(year==2007, continent == c("Europe", "Africa"), lifeExp>60)
```

```
# Add a geom_text() layer
plot_pst <- ggplot(gm2007, aes(x = lifeExp, y = country, color = lifeExp)) +
  geom_point(size = 10) +
  geom_segment(aes(xend = 50, yend = country), size = 4) +
  geom_text(aes(label = round(lifeExp)), color = "white", size = 4)

plot_pst
```

### change to theme_classic()

```
plot_pst +
  theme_classic()
```

### Calculate mean and plot as a vertical line

```
# Calculate mean of lifeExp for couuntries in dataset
global_mean_filtered_countries <- mean(gm2007$lifeExp)
```

```
# Plot with vertical line
plot_pst +
  theme_classic() +
  geom_vline(xintercept = global_mean_filtered_countries, color = "grey40", linetype = 3)
```

### Plot an arrow

```
x_start <- global_mean_filtered_countries + 8
y_start <- 10.5
x_end <- global_mean_filtered_countries
y_end <- 11
```

```
 # Plot with vertical line and arrow
 
 plot_pst +
  theme_classic() +
  geom_vline(xintercept = global_mean_filtered_countries, color = "grey40", linetype = 3) +
  annotate("curve", x = x_start, y = y_start,
    xend = x_end, yend = y_end,
    arrow = arrow(length = unit(0.2, "cm"), type = "closed"),
    color = "grey40"
  )
```

### Plot an annotation with arrow

```
 # Plot with vertical line and arrow
 
 plot_pst +
  theme_classic() +
  geom_vline(xintercept = global_mean_filtered_countries, color = "grey40", linetype = 3) +
  annotate("curve", x = x_start, y = y_start, xend = x_end, yend = y_end,
    arrow = arrow(length = unit(0.2, "cm"), type = "closed"), color = "grey40") +
  annotate("text",x = x_start, y = y_start, label = "The\nglobal\naverage",
    vjust = 1, size = 3, color = "grey40") + 
    labs(title="Awesome point with segment plot! ")
```
